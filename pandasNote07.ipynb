# 日付、時刻の扱い

## 基本的なデータ型（クラス）

```python
d2 = pd.Timestamp.now(tz='Asia/Tokyo') # tz -> タイムゾーン
print(d2)
```

### 部分情報の取り出し

```python
print(d2.year) # 年のプロパティ（変数）を取得する
print(d2.month)
print(d2.day)
print(d2.hour)
print(d2.minute)
print(d2.second)
print(d2.microsecond)
print(d2.weekday()) # 曜日（weekday）は、メソッド（関数）
```

### 指定した日付、時刻の情報を作る

```python
d3 = pd.Timestamp('1966-03-14T13:15:00+09:00') # UTCとの差を指定する形式
print(d3)
```

```python
d3 = pd.Timestamp('1966-03-14T13:15:00', tz='Asia/Tokyo') # 地域名で指定する方法
print(d3)
```

```python
d3 = pd.Timestamp("19660314131500", tz='Asia/Tokyo')
print(d3)
```

#### UTCで日付、時刻を作成

```python
d3 = pd.Timestamp("1966-03-14T04:15:00Z")
print(d3)
```

### タイムゾーンの変換

```python
d3.tz_convert('Asia/Tokyo')
```

## タイムスタンプの差

```python
d1 = pd.Timestamp('2000-01-01T00:00:00', tz='Asia/Tokyo')
d2 = pd.Timestamp('2022-03-14T13:15:00', tz='Asia/Tokyo')
td21 = d2 - d1
print(td21)
```

### 未来の日付、時刻の算出

```python
d2 + td21
```

### Timedeltaオブジェクトの作成

```python
td100 = pd.Timedelta('100 days 01:20:30')
td100
```

```python
td100 = pd.Timedelta(days=100, hours=1, minutes=20, seconds=30)
td100
```

## 応用

### 日付、時刻の系列を作成する方法

```python
d1 = pd.Timestamp('2022-01-01T00:00:00')
d2 = pd.Timestamp('2022-12-31T23:59:59')
```

```python
# 1か月間隔の1日の日付の並び
dr = pd.date_range(d1, d2, freq='MS') # MS -> 月ごとに取得する
dr
```

```python
# 1日間隔の日付の並び
dr = pd.date_range(d1, d2, freq='D') # MS -> 月ごとに取得する
dr
```
