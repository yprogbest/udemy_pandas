```python
import pandas as pd
```

# DataFrameをCSVファイルとして保存する

```python
lst = [[1,1,1], [2,4,8], [3,9,27]]
df = pd.DataFrame(lst, index=['d1','d2','d3'], columns=['linear', 'square', 'cubic'])
df
```

## Notebookのディレクトリを取得する

```python
dname = r'/content/drive/MyDrive/Colab Notebooks'
dname
```

```python
fname = dname + r'/csv01.csv'
fname
```

## DataFrameの内容をCSVファイルに保存する

```python
df.to_csv(fname)
```

```python
fname2 = dname + r'/csv01_noindex.csv'
fname2
```

```python
df.to_csv(fname2, index=False) # インデックスの列を省略するケース
```

```python
fname3 = dname + r'/csv01_nohead.csv'
fname3
```

```python
df.to_csv(fname3, header=False) # 見出し行を省略するケース
```

## CSVファイルの内容をDataFrameに読み込む

```python
df2 = pd.read_csv(fname,index_col=0)
df2
```

```python
df2 = pd.read_csv(fname2) # インデックスがない
df2
```

```python
df2 = pd.read_csv(fname3, index_col=0, names=['linear', 'square', 'cubic']) # ヘッダ行をつける
df2
```

```python
fname4 = dname + r'/file_sjis.csv'
df3 = pd.read_csv(fname4, encoding='sjis', index_col=0)
df3
```
